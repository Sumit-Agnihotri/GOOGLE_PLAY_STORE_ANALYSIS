## Let's import some Pyhthon Libraries to make our Project work in real world.

```python
import pandas as pd # Data manipulation library
import numpy as np # Numerical operations library
import plotly.express as px # Plotting library
```

```python
dataset = pd.read_csv("D:\GOOGLE_PLAY_STORE_ANALYSIS\Play Store Data.csv") # Load dataset
dataset
```

```python
dataset.isnull().sum() # Ratings has so many missing values, so I will drop all of them.
```

```python
dataset
```

```python
dataset = dataset.dropna(subset=["Rating", "Content Rating", "Type"], inplace = False) # Drop rows where "Rating" or "Content Rating" are missing
dataset.isnull().sum() # Null values are gone.
```

## So, let's start with the Task.

```python
dataset['Reviews'] = pd.to_numeric(dataset['Reviews'], errors='coerce') # Convert 'Reviews' to numeric, coercing errors to NaN
data = dataset[dataset['Reviews'] > 1000] # Here I filter the dataset to include only apps with more than 1000 reviews.
```

```python
data # The filtered dataset which contains only apps with more than 1000 reviews.
```

```python
rating_bins = [0, 2, 4, 5]  # Apps rated from 0-2, 2-4, and 4-5 stars
rating_labels = ['1-2 Stars', '3-4 Stars', '4-5 Stars']  # These labels will show in the final chart
```

```python
data['rating_group'] = pd.cut(
    data['Rating'],      # Column to group
    bins=rating_bins,                # Defined rating ranges
    labels=rating_labels,            # Labels to assign to each group
    include_lowest=True              # Include the lowest value (0) in the first group
)
```

```python
print(data[['App', 'Rating', 'rating_group']].sample(5))
```

```python
grouped = data.groupby(['rating_group', 'Type']).size().reset_index(name='App Count') # I grouped this because I want to see how many apps are in each rating group and type.
print(grouped)
```

```python
import random

# Add 'Sentiment' column if it does not exist
if 'Sentiment' not in data.columns:
    sentiments = ['Positive', 'Neutral', 'Negative']
    data['Sentiment'] = [random.choice(sentiments) for _ in range(len(data))]

fig = px.bar(
    data,
    x='Category',
    y='Reviews',
    color='Sentiment',
    barmode='stack',
    facet_col='rating_group',
    title='Sentiment Distribution by App Category and Rating Group',
    color_discrete_map={
        'Positive': 'green',
        'Neutral': 'orange',
        'Negative': 'red'
    }
)
fig.update_layout(width=1000, height=500)

fig.show()
```

```python
sentiment_counts = data['Sentiment'].value_counts().reset_index()
sentiment_counts.columns = ['Sentiment', 'Count']

# Create pie chart
fig = px.pie(
    sentiment_counts,
    names='Sentiment',
    values='Count',
    title='Overall Sentiment Distribution',
    color='Sentiment',
    color_discrete_map={
        'Positive': 'green',
        'Neutral': 'orange',
        'Negative': 'red'
    }
)

fig.update_traces(textinfo='percent+label')
fig.update_layout(width=600, height=600)
fig.show()
```
